```python
def makeSquareGraph(n):
    #TODO ADD edge costs
    g_grid = nx.grid_graph(dim=(n,n))
    return nx.convert_node_labels_to_integers(g_grid)
```

```python
g = makeSquareGraph(5)
edge = EdgeCostLinear(6,2,3,10)
costly_edge = EdgeCostLinear(6,2,3,10)
cheap_edge = EdgeCostLinear(10,6,1,8)
#TODO Subgraph these 
graphEdgeFunc("Edge for grid", cheap_edge)
graphEdgeFunc("Edge for path", costly_edge)
```

```python
g_func = CostFuncGraph(g)
print(f"Made graph without functions, should be invalid: {g_func.is_valid()}")
```

```python
g_func.set_functions(g_func.edges(), cheap_edge)
    
#This is hardcoded for now...
costly_path = [(0, 5), (5, 6), (6, 11), (11, 16), (16, 21), (21, 22), (22, 23), (23, 24)]

g_func.set_functions(costly_path, costly_edge)

print(f"Added functions, should be valid now: {g_func.is_valid()}")
```

```python
g_dot = nx.drawing.nx_pydot.to_pydot(g)
img = g_dot.create_svg(prog='dot')
display(SVG(img))
```

```python
""" Model settings """
start_node = 0
end_node = 24
B = 20
n = 300
```

```python
def change_budget(b):
    model = Model(g_func, start_node, end_node, b, n, verbose=True)
    s_path = model.solve(print_path=True)
    saveGraphImgs(g_func,s_path);
```

```python
interact_opt = interact_manual.options(manual_name="Run optimization")
interact_opt(change_budget, b=widgets.IntSlider(min=10, max=60, step=1, value=20));
```
